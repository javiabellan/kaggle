```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data

```python
train = pd.read_csv('data/train_clean.csv', dtype={'time': np.float32, 'signal': np.float32, 'open_channels':np.int32})
test  = pd.read_csv('data/test_clean.csv',  dtype={'time': np.float32, 'signal': np.float32})
sub   = pd.read_csv('data/sample_submission.csv', dtype={'time': np.float32})
```

```python
train_groups = np.concatenate((
    np.repeat([0],1000000),
    np.repeat([1], 500000),
    np.repeat([2], 500000),
    np.repeat([3], 500000),
    np.repeat([4], 500000),
    np.repeat([1], 500000),
    np.repeat([2], 500000),
    np.repeat([4], 500000),
    np.repeat([3], 500000)))#.astype(np.int8)

test_groups = np.concatenate((
    np.repeat([0], 100000), 
    np.repeat([2], 100000),
    np.repeat([4], 100000),
    np.repeat([0], 100000),
    np.repeat([1], 100000),
    np.repeat([3], 100000),
    np.repeat([4], 100000),
    np.repeat([3], 100000),
    np.repeat([0], 100000),
    np.repeat([2], 100000),
    np.repeat([0],1000000)))#.astype(np.int8)


len(train_groups), len(test_groups)
```

# Guardar

```python
np.save('data/0_groups_train.npy', train_groups)
np.save('data/0_groups_test.npy',  test_groups)
```
